### Wykresy rozkadw - joinplot

```python
sns.set(style = 'darkgrid')
```

```python
df3 = np.random.multivariate_normal(mean = [0, 1], cov = [(1, -0.4), (-0.4, 1)], size = 200)
df3 = pd.DataFrame(df3, columns = ['x', 'y'])
df3.head()
```

```python
df3.describe()
```

```python
sns.jointplot(data = df3, x = 'x', y = 'y')#x - rozklad normalny standardowy, y = wart oczekiwana 1, rozszrut 1
```

```python
sns.set(style = 'white')
sns.jointplot(data = df3, x = 'x', y = 'y', kind = 'hex')
```

```python
sns.jointplot(data = df3, x = 'x', y = 'y', kind = 'kde')
```

```python
sns.jointplot(data = df3, x = 'x', y = 'y', kind = 'kde', n_levels=15)
```

```python
g = sns.jointplot(data = df3, x = 'x', y = 'y', kind = 'kde', n_levels=15, cmap = 'viridis')
g
```
